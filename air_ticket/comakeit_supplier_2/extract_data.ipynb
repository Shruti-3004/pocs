```python
import os
import glob
from prettyprinter import pprint
import datetime

def get_supplier_name(text):
    for line in text:
        if 'paytm' in line.lower():
            return ['PAYTM']
        if 'amazon' in line.lower():
            return ['AMAZON']

def get_item_names(text):
    supplier = get_supplier_name(text)
    for i in range(len(text)):
        if 'postpaid mobile' in text[i].lower():
            if supplier[0].lower() == 'amazon':
                return ['Amazon Bill- Post paid Mobile Bill']
        if 'water bill' in text[i].lower():
            if supplier[0].lower() == 'amazon':
                return ['Amazon Bill- Water bill']
        if 'landline bill' in text[i].lower():
            if supplier[0].lower() == 'amazon':
                return ['Amazon Bill- Landline bill']
        if 'description' in text[i].lower():
            item = ''
            for j in range(i+1, len(text)):
                if 'total' in text[j].lower():
                    return [item]
                words = text[j].split()
                if 'Rs.' in words:
                    ind = words.index('Rs.')
                    item += ' '.join(words[:ind])
                else:
                    item += text[j]

def get_spent_amount(text):
    for i in range(len(text)):
        if 'grand total' in text[i].lower():
            return [text[i].split()[-1]]
        if 'amount' in text[i].lower():
            return [text[i+1].split()[-1]]  

def get_currency_code(text):
    for line in text:
        if 'rs' or 'inr' in line.lower():
            return ['INR']     

def get_spending_category(text):
    lineitem = get_item_names(text)
    if lineitem is not None:
        lineitem = lineitem[0]

    if lineitem is not None:
    
        if 'landline' in lineitem.lower() or 'fibernet' in lineitem.lower():
            return ['Landline Services']
        if 'water' in lineitem.lower():
            return ['Water Services']
        if 'mobile' in lineitem.lower() or 'airtel' in lineitem.lower() or 'jio' in lineitem.lower():
            return ['Mobile Services']
        
def get_spent_date(text):
    t = ' '.join(text)
    print(t)
    for line in text:
        if 'bill date' in line.lower():
            date = line.split()[-1]
            break
        if 'order placed' in line.lower():
            date = ' '.join(line.split()[-3:])
            break
    formats = ['%d-%m-%Y', '%d %B %Y']
    for format in formats:
                try:
                    d = datetime.datetime.strptime(date, format).strftime('%Y-%m-%d')
                    return [d]
                except:
                    pass

def get_quantity(text):
    items = get_item_names(text)
    if items is None:
        return [0]
    return [len(items)]

def get_unit_of_measure(text):
    return ['Each']

for file in glob.glob('C:/air_ticket/comakeit_supplier_2/output/*.txt'):
    print(file)
    filename = os.path.split(file)[-1]

    stationary_asset_final = []
    spent_amount_final = []
    currency_code_final = []
    spent_date_final = []
    supplier_name_final = []
    item_names_final = []
    quantities_final = []
    spending_category_final = []
    line_item_amounts_final = []
    unit_of_measure_final = []
    
    with open(file) as fp:

        texts = fp.read().split("---\n")
        # pprint(texts)
        for text in texts:
            text = text.splitlines()
            text = [line for line in text if not line.isspace() and len(line) > 0]
            # pprint(text)
            filename = os.path.split(file)[-1].split('.txt')[0]

            supplier_name = get_supplier_name(text)
            if supplier_name and not supplier_name_final:
                supplier_name_final = supplier_name

            item_names = get_item_names(text)
            if item_names and not item_names_final:
                item_names_final = item_names

            spent_amount = get_spent_amount(text)
            if spent_amount and not spent_amount_final:
                spent_amount_final = spent_amount

            currency_code = get_currency_code(text)
            if currency_code and not currency_code_final:
                currency_code_final = currency_code

            spending_category = get_spending_category(text)
            if spending_category and not spending_category_final:
                spending_category_final = spending_category

            spent_date = get_spent_date(text)
            if spent_date and not spent_date_final:
                spent_date_final = spent_date

            quantity = get_quantity(text)
            if quantity and not quantities_final:
                quantities_final = quantity

            unit_of_measure = get_unit_of_measure(text)
            if unit_of_measure and not unit_of_measure_final:
                unit_of_measure_final = unit_of_measure
    
    print(supplier_name_final)
    print(item_names_final)
    print(spent_amount_final)
    print(currency_code_final)
    print(spending_category_final)
    print(spent_date_final)
    print(quantities_final)
    print(unit_of_measure_final)
    print('-----------------------------------------')

```
