```python
import glob
import os
import re

def get_stationary_asset(text):
    asset = []

    data = {
        'first': 'Dendukuri House',
        'second': 'Galaxy Aurobindo 9th Floor',
        'third': 'Galaxy Aurobindo 5th Floor',
    }

    for line in text:
        if re.search('.*plot no.*564/.*a.*39.*', line.lower()):
            asset.append(data['first'])
            break
    
    return asset

def get_consumption_unit(text):
    return ['KWH']

def isfloat(num):
    try:
        float(num)
        return True
    except ValueError:
        return False

def get_fuel_consumption(text):
    cons = []
    flag = 0
    for line in text:
        if 'total consumption' in line.lower() or 'otal consumption' in line.lower():
            words = line.split()
            for i in range(len(words)-1):
                if words[i].lower() == 'consumption':
                    # print('words---> ', words)
                    words[i+1] = words[i+1].strip('-_—__')
                    if isfloat(words[i+1]) and '.' in words[i+1] and len(words[i+1].split('.')[-1]) > 0 and len(words[i+1].split('.')[-1]) < 3:
                            flag = 1
                            cons.append(words[i+1])
                            break
                if flag:
                    break
        if flag:
            break
    return cons

def get_supplier(text):
    supplier = []
    for line in text:
        if 'limited' in line.lower():
            supplier.append(line)
            break
    return supplier

def get_date(text):
    start_date = []
    end_date = []
    for line in text:
        if 'date' in line.lower():
            words = line.split()
            if '-' in words[-1] and not re.search('[a-zA-Z]', words[-1]):
                # print(words[-1])
                month = words[-1].split('-')[1]
                if month == '01':
                    start_date.append('2022-01-01')
                    end_date.append('2022-01-31')
                    break
                if month == '02':
                    start_date.append('2022-02-01')
                    end_date.append('2022-02-28')
                    break
                if month == '03':
                    start_date.append('2022-03-01')
                    end_date.append('2022-03-31')
                    break
                if month == '04':
                    start_date.append('2022-04-01')
                    end_date.append('2022-04-30')
                    break
                if month == '05':
                    start_date.append('2022-05-01')
                    end_date.append('2022-05-31')
                    break
                if month == '06':
                    start_date.append('2022-06-01')
                    end_date.append('2022-06-30')
                    break
                if month == '07':
                    start_date.append('2022-07-01')
                    end_date.append('2022-07-31')
                    break
                if month == '08':
                    start_date.append('2022-08-01')
                    end_date.append('2022-08-31')
                    break
                if month == '09':
                    start_date.append('2022-09-01')
                    end_date.append('2022-09-30')
                    break
                if month == '10':
                    start_date.append('2022-10-01')
                    end_date.append('2022-10-31')
                    break
                if month == '11':
                    start_date.append('2022-11-01')
                    end_date.append('2022-11-30')
                    break
                if month == '12':
                    start_date.append('2022-12-01')
                    end_date.append('2022-12-31')
                    break
                # break
    return start_date, end_date

def get_fuel_type(text):
    return ['Electricity']

for file in glob.glob('C:\\air_ticket\\comakeit_electricity\\ensemble_files\\*.txt'):
    print(file)
    # files.append(file)
    filename = os.path.split(file)[-1]

    stationary_asset_final = []
    fuel_consumption_final = []
    consumption_unit_final = []
    supplier_final = []
    start_date_final = []
    end_date_final = []
    fuel_type_final = []
    
    with open(file) as fp:

        texts = fp.read().split("---\n")
        # pprint(texts)
        for text in texts:
            text = text.splitlines()
            text = [line for line in text if not line.isspace() and len(line) > 0]
            # pprint(text)
            filename = os.path.split(file)[-1].split('.txt')[0]

            stationary_asset = get_stationary_asset(text)
            if stationary_asset and not stationary_asset_final:
                stationary_asset_final = stationary_asset

            fuel_consumption = get_fuel_consumption(text)
            if fuel_consumption and not fuel_consumption_final:
                fuel_consumption_final = fuel_consumption

            consumption_unit = get_consumption_unit(text)
            if consumption_unit and not consumption_unit_final:
                consumption_unit_final = consumption_unit
            
            supplier = get_supplier(text)
            if supplier and not supplier_final:
                supplier_final = supplier

            start_date = get_date(text)[0]
            if start_date and not start_date_final:
                start_date_final = start_date

            end_date = get_date(text)[1]
            if end_date and not end_date_final:
                end_date_final = end_date

            fuel_type = get_fuel_type(text)
            if fuel_type and not fuel_type_final:
                fuel_type_final = fuel_type

    if not stationary_asset_final:
        stationary_asset_final = [None]
    if not fuel_consumption_final:
        fuel_consumption_final = [None]
    if not consumption_unit_final:
        consumption_unit_final = [None]
    if not supplier_final:
        supplier_final = [None]
    if not start_date_final:
        start_date_final = [None]
    if not end_date_final:
        end_date_final = [None]
    if not fuel_type_final:
        fuel_type_final = [None]
            
    print('STATIONARY ASSET---> ', stationary_asset_final)
    print('CONSUMPTION---> ', fuel_consumption_final)
    print('UNIT---> ', consumption_unit_final)
    print('SUPPLIER---> ', supplier_final)
    print('START DATE---> ', start_date_final)
    print('END DATE---> ', end_date_final)
    print('FUEL TYPE---> ', fuel_type_final)
    print('-------------------------------------------')
```
