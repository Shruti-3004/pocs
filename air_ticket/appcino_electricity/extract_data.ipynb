```python
import glob
import os
from prettyprinter import pprint
import datetime

def get_stationary_asset(text):
    pprint(text)

def get_start_date(text):
    start_date = ''
    for line in text:
        if 'consumption period' in line.lower():
            for word in line.split():
                ## add date condition below
                if '/' in word:
                    formats = ['%d/%m/%Y']
                    for format in formats:
                        try:
                            start_date = datetime.datetime.strptime(word, format).strftime('%Y-%m-%d')
                            return start_date
                        except:
                            pass

def get_end_date(text):
    count = 0
    end_date = ''
    for line in text:
        if 'consumption period' in line.lower():
            for word in line.split():
                ## add date condition below
                if '/' in word:
                    if count:
                        formats = ['%d/%m/%Y']
                        for format in formats:
                            try:
                                end_date = datetime.datetime.strptime(word, format).strftime('%Y-%m-%d')
                                return end_date
                            except:
                                pass
                    count += 1

def get_fuel_consumption(text):
    flag = 0
    for i in range(len(text)):
        if 'consumption unit' in text[i].lower():
            amount = 0
            for j in range(i+1, len(text)):
                if 'total' in text[j].lower():
                    flag = 1
                    break
                words = text[j].split()
                for word in words:
                    if not word.isalpha():
                        amount += float(word)
                        break
        if flag:
            return round(amount, 2)

def get_consumption_unit(text):
    return 'kwh'

def get_fuel_type(text):
    return 'electricity'

def get_supplier(text):
    return 'rajasthan patrika pvt ltd'

def get_stationary_asset(text):
    return 'appcino 4th floor and 2nd floor'

for file in glob.glob('C:/air_ticket/appcino_electricity/ensemble_files/*.txt'):
    print(file)
    # files.append(file)
    # filename = os.path.split(file)[-1]
    filename = os.path.split(file)[-1].split('.txt')[0]
    # print(filename)
    f = filename.split('_')
    ind = 0
    for i in range(len(f)):
        if 'page' in f[i]:
            ind = i
            break
    filename_final = '_'.join(f[:ind])
    # print(filename_final)

    stationary_asset_final = []
    fuel_consumption_final = []
    consumption_unit_final = []
    supplier_final = []
    start_date_final = []
    end_date_final = []
    fuel_type_final = []


    
    with open(file, errors='ignore') as fp:

        texts = fp.read().split("---\n")
        # pprint(texts)
        for text in texts:
            text = text.splitlines()
            text = [line for line in text if not line.isspace() and len(line) > 0]
            # pprint(text)
            # filename = os.path.split(file)[-1].split('.txt')[0]

            stationary_asset = get_stationary_asset(text)
            if stationary_asset and not stationary_asset_final:
                stationary_asset_final = stationary_asset

            start_date = get_start_date(text)
            if start_date and not start_date_final:
                start_date_final = start_date

            end_date = get_end_date(text)
            if end_date and not end_date_final:
                end_date_final = end_date

            fuel_consumption = get_fuel_consumption(text)
            if fuel_consumption and not fuel_consumption_final:
                fuel_consumption_final = fuel_consumption

            consumption_unit = get_consumption_unit(text)
            if consumption_unit and not consumption_unit_final and fuel_consumption_final:
                consumption_unit_final = consumption_unit
            
            fuel_type = get_fuel_type(text)
            if fuel_type and not fuel_type_final and fuel_consumption_final:
                fuel_type_final = fuel_type

            supplier = get_supplier(text)
            if supplier and not supplier_final:
                supplier_final = supplier

    print('STATIONARY ASSET ---> ', stationary_asset_final)
    print('START DATE ---> ', start_date_final)
    print('END DATE ---> ', end_date_final)
    print('FUEL CONSUMPTION ---> ', fuel_consumption_final)
    print('CONSUMPTION UNIT ---> ', consumption_unit_final)
    print('FUEL TYPE ---> ', fuel_type_final)
    print('SUPPLIER ---> ', supplier_final)
```
