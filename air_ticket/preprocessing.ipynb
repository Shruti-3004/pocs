```python
import cv2
import os

image = 'C:\\air_ticket\\delimiters2\\Yatra_AAAIN233676183_20.09.2022_page0.jpg'
img = cv2.imread(image)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

k = 5
result = 255*(img/255)**k

cv2.imwrite('result.jpg', result)

command = 'tesseract result.jpg stdout -l eng --psm 4'
print(os.popen(command).read())

```
