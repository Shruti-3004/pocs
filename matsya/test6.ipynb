```python
traveler_names = list(data['Traveler Name'])
# print(traveler_names)

new_traveler_names = []
for tname in traveler_names:
    words = tname.split()
    if len(words) > 2:
        if 'mr' in words[0].lower() or 'ms' in words[0].lower():
            new_traveler_names.append(' '.join(words[1:3]))
        else:
            new_traveler_names.append(' '.join(words[:3]))
    else:
        if 'mr' in words[0].lower() or 'ms' in words[0].lower():
            new_traveler_names.append(' '.join(words[1:]))
        else:
            new_traveler_names.append(' '.join(words))

print(len(new_traveler_names))
```

```python
data['New Traveler Name'] = new_traveler_names
```
