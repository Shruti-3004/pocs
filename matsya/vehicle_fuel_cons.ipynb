```python
import pandas as pd
```

```python
df = pd.read_excel('C:\sample_invoice_output\matsya\Vehicle Asset Energy Consumption (1).xlsx')
```

```python
df
```

```python
from datetime import datetime

def months_between_dates(start_date, end_date):
    months = (end_date.year - start_date.year) * 12 + end_date.month - start_date.month
    print('months---> ', months)
    return months
```

```python
df['Number of Months'] = 0
```

```python
for i in range(len(df)):
    df['Number of Months'].iloc[i] = months_between_dates(df['Start Date'].iloc[i], df['End Date'].iloc[i])
    # print(df['Start Date'].iloc[i])
```

```python
df
```

```python
df['Fuel consumption for 2023'] = 0
df['KwH consumption for 2023'] = 0
```

```python
for i in range(len(df)):
    if df['Number of Months'].iloc[i] > 12:
        df['Fuel consumption for 2023'].iloc[i] = (df['Fuel consumption'].iloc[i]/df['Number of Months'].iloc[i])*12
        df['KwH consumption for 2023'].iloc[i] = (df['KwH consumption'].iloc[i]/df['Number of Months'].iloc[i])*12
    else:
        df['Fuel consumption for 2023'].iloc[i] = df['Fuel consumption'].iloc[i]
        df['KwH consumption for 2023'].iloc[i] = df['KwH consumption'].iloc[i]
```
